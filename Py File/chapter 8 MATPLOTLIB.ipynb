```python
import matplotlib.pyplot as plt
import numpy as np
x = np.random.normal(170,10,250)
print(x)
plt.hist(x, color='r')
plt.show()
```

```python
fruits = np.array([25,15,30,30])
x = np.array(['banana','mango','guava','pineapple'])
plt.pie(fruits,labels = x, startangle = 90, autopct = '%1.1f%%')
plt.show()
```

```python
fruits = np.array([25,15,30,30])
x = np.array(['banana','mango','guava','pineapple'])
myexplode = [0.2,0,0.1,0]
plt.pie(fruits,labels = x,explode = myexplode,shadow = True, startangle = 90, autopct = '%1.1f%%')
plt.show()
```

```python
#100 students in classs in which 25 students are failed, 25 are in distinction,25 are in first class and rest are in 
#second class.
x = np.array([25,25,25,25])
y = np.array(['dist','first','second','fail'])
myexplode = np.array([0.2,0.2,0,0.1])
plt.pie(x,labels = y,explode = myexplode,shadow = True,startangle = 90, autopct = '%1.1f%%')
plt.show()
```

```python
x = np.array([25,25,25,25])
y = np.array(['dist','first','second','fail'])
myexplode = np.array([0.2,0.2,0,0.1])
mycolors = ['red','yellow','b','#4CAF50']
plt.pie(x,labels = y,explode = myexplode,colors = mycolors,shadow = True,startangle = 90, autopct = '%1.1f%%')
plt.show()
```

```python
fruits = np.array([25,15,30,30])
x = np.array(['banana','mango','guava','pineapple'])
plt.pie(fruits,labels = x, startangle = 90, autopct = '%1.1f%%')
plt.legend(title = "Fruits", loc = 'lower right')
plt.show()
```
